# **Import library and data**

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/kaggle/input/taylor-concert-tours-impact-on-attendance-and/Taylor_Train.csv", encoding="latin-1")
df.describe(include="all")
df.info()
df.shape
df.head(5).T
df.isnull().sum()

warnings.filterwarnings("ignore")
```

# **Visualize the dataset******

```python
sns.heatmap(df.isna())
```

```python
Country = df['Country'].value_counts()
Country
plt.figure(figsize=(20, 15))
plt.xticks(rotation=70, fontsize=10)
sns.set_palette("colorblind")
sns.barplot(x=Country.index, y=Country.values)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Count of each Country')
```

```python
Venue = df['Venue'].value_counts()
Venue
plt.figure(figsize=(20, 15))
plt.xticks(rotation=70, fontsize=10)
sns.set_palette("colorblind")
sns.barplot(x=Venue.index, y=Venue.values)
plt.xlabel('Venue')
plt.ylabel('Count')
plt.title('Count of each Venue')
```
